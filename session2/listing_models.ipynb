# Listing models
This code can be used to list the models available on Helmholtz' [Blablador LLM server](https://blablador.fz-juelich.de/).

```python
from dotenv import load_dotenv
load_dotenv()
import openai 
import os
client = openai.OpenAI()
client.base_url = "https://api.helmholtz-blablador.fz-juelich.de/v1"
client.api_key = os.environ.get('BLABLADOR_API_KEY')

print("\n".join([model.id for model in client.models.list().data]))
```
